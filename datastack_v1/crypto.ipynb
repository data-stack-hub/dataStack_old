```python
#  profit 
pl_doc = db.collection('spaces').document('a9ES0AGH6GWlec8hnNqA').collection('states').get()
pl_data = []
for doc in pl_doc:
    acc = []
    for key, value in doc.to_dict()['days'].items():
        value['day'] = key + "-2022"
        acc.append(value)
    pl_data = pl_data + acc
pl_data = pd.DataFrame(pl_data)
pl_data = pl_data.sort_values(by=['day'])
pl_data['cumm_pnl'] = pl_data.pnl.cumsum()

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(
    x=pl_data['day'],
    y=pl_data['cumm_pnl'],
    name = 'pnl', # Style name/legend entry with html tags
    connectgaps=True, # override default to connect the gaps
    
))

print('total : ' ,pl_data['pnl'].sum())
fig.show()

```
